```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import string
import seaborn as sns
df2 = pd.read_csv("C:/research/step4-file.csv")
df1 = pd.read_csv("C:/research/compare_thresholds_res.csv")

plt.title("accurancy(RF-distance) comparison between different thresholds")
plt.ylabel("RF-distance")
t1 = df1[(df1["THRES"] == 0.5)]
t2 = df1[(df1["THRES"] == 0.8)]
t3 = df1[(df1["THRES"] == 0.9)]
B = [t1["RF"],t2["RF"],t3["RF"]]
plt.boxplot(B, patch_artist=True, positions = [1,2,3] , widths = 0.6)
plt.xticks([1,2,3],["0.5","0.8","0.9"])
plt.figure()
```

```python
plt.title("running time comparison between different thresholds")
plt.ylabel("running time")
r1 = df2[(df2["threshold"] == 0.5)]
r2 = df2[(df2["threshold"] == 0.8)]
r3 = df2[(df2["threshold"] == 0.9)]
B = [r1["time"],r2["time"],r3["time"]]
plt.boxplot(B, patch_artist=True, positions = [1,2,3] , widths = 0.6)
plt.xticks([1,2,3],["0.5","0.8","0.9"])
plt.figure()
```

```python
plt.title("ML score comparison between different thresholds")
plt.ylabel("ML score")
m1 = df2[(df2["threshold"] == 0.5)]
m2 = df2[(df2["threshold"] == 0.8)]
m3 = df2[(df2["threshold"] == 0.9)]
B = [m1["loglikelihood"],m2["loglikelihood"],m3["loglikelihood"]]
plt.boxplot(B, patch_artist=True, positions = [1,2,3] , widths = 0.6)
plt.xticks([1,2,3],["0.5","0.8","0.9"])
plt.figure()
```
